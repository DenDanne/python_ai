### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
Det finns
* Object-oriented style:
Detta sätt är det rekommenderade då det ger mer kontroll, flexibilitet och tydlighet.
Detta eftersom man explicit skapar de olika objekten och sedan kallar methods på dem för att skapa plottar.

* Pyplot implicit:
Här ger man pyplot ansvaret för att hantera figurer och axises. 
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
Figure: Det är delen som håller alla delar av matplotlib-plotten. 
Axes: Innehåller den faktiska datan
Artist: Linjer, markörer, text osv.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
Den främsta input data-type som förväntas är numpy.arrays.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
fig, ax = plt.subplots()
x = np.linspace(-4, 4, 50)
y = x**2

ax.plot(x, y, label="y=x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.title("y=x^2")
plt.legend()
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.scatter(random_data_x, random_data_y)
ax1.set_title("scatterplot of random data")
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2.bar(names, values)
ax2.set_title("Nr fruites")
ax2.set_xlabel("Type fruit")
ax2
```
